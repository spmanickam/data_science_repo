## Mini Project: Car & Pedestrian Detection

```python
# Pedestian Detection
import cv2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import Image 

import time

pretrained_body_classifier = cv2.CascadeClassifier('../../models/pre-trained/Haarcascades/haarcascade_fullbody.xml')
capture = cv2.VideoCapture('../../data/raw/U_CV_RATAN/walking.avi')

while capture.isOpened():
    time.sleep(0.1)  #high value reduces playback speed
    ret, frame = capture.read()
    # Reduce the frame size to speed up things
    frame = cv2.resize(frame, None, fx = 0.5, fy = 0.5, interpolation = cv2.INTER_LINEAR)
    
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Scaling parameter (Ex: 1.3) -> Larger values increase the detection faster
    # Sensitivity factor -> Number of bounding boxes to confirm object detection( smaller values leads to more Flase positives)
    bodies_detected = pretrained_body_classifier.detectMultiScale(frame_gray, 1.3, 1)
    
    for (x, y, w, h) in bodies_detected:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255, 0, 0), 2)
        cv2.imshow('Pedestrians detected: ', frame)
    if cv2.waitKey(1) == 13:
        break
capture.release()   
cv2.destroyAllWindows()
```

```python
# Car Detection
import cv2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import Image 

import time

pretrained_body_classifier = cv2.CascadeClassifier('../../models/pre-trained/Haarcascades/haarcascade_car.xml')
capture = cv2.VideoCapture('../../data/raw/U_CV_RATAN/cars.avi')

while capture.isOpened():
    time.sleep(0.1)  #high value reduces playback speed
    ret, frame = capture.read()
    # Reduce the frame size to speed up things
    #frame = cv2.resize(frame, None, fx = 2, fy = 2, interpolation = cv2.INTER_LINEAR)
    
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Scaling parameter (Ex: 1.3) -> Larger values increase the detection faster
    # Sensitivity factor -> Number of bounding boxes to confirm object detection( smaller values leads to more Flase positives)
    bodies_detected = pretrained_body_classifier.detectMultiScale(frame_gray, 1.3, 1)
    
    for (x, y, w, h) in bodies_detected:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255, 0, 0), 2)
        cv2.imshow('CARs detected: ', frame)
    if cv2.waitKey(1) == 13:
        break
capture.release()   
cv2.destroyAllWindows()
```
