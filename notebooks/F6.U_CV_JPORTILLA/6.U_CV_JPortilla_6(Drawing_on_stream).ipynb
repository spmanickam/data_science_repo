## Using video files

```python
import cv2
import time

cap = cv2.VideoCapture('my_capture_video.mp4')

if not cap.isOpened():
    print('Error reading the file or wrong CODEC used!')

while cap.isOpened():
    return_value, frame = cap.read()
    
    if return_value:
        
        time.sleep(1/20)  # use the frame in denominator for slowing the speed
        cv2.imshow('frame', frame)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break
```

## Drawing on a video stream

```python
import cv2

cap = cv2.VideoCapture(0)

width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

#width, height #(1280.0, 720.0)

x = int(width // 2)
y = int(height // 2)

#width of rectangle
w = int(width // 4)
h = int(height // 4)

print (x,y,w,h)

while True:
    return_value, frame = cap.read()
    cv2.rectangle(frame, (x,y), (x + w, y + h), color = (255, 255, 0), thickness = 10)
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

def draw_rectangle(event, x, y , flags, param):
    global point1, point2, top_left_clicked, bottom_right_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
         
        #RESET
        if top_left_clicked and bottom_right_clicked:
            point1 = (0,0)
            point2 = (0,0)
            top_left_clicked = False
            bottom_right_clicked = False
        
        if top_left_clicked == False:
            point1 = (x,y)
            top_left_clicked = True
            print(point1)
        
        elif bottom_right_clicked == False:
            point2 = (x,y)
            bottom_right_clicked = True
            print(point2)

# global variables:
point1 = (0,0)
point2 = (0,0)
top_left_clicked = False
bottom_right_clicked = False

cv2.namedWindow('Test')
cv2.setMouseCallback('Test', draw_rectangle)

while True:
    
    return_value, frame = cap.read()
    print('### 1 ###')
    print('top_left_clicked:', top_left_clicked)
    print('bottom_right_clicked:', bottom_right_clicked)
    
    if top_left_clicked:
        cv2.circle(frame, point1, 5, color = [255, 0, 255], thickness=-1)
        print('### 2 ###')
        
    if top_left_clicked and bottom_right_clicked:
        cv2.rectangle(frame, point1 , point2, color = [255, 0, 0], thickness=5)
        print('### 3 ###')
        
    cv2.imshow('Test', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
