```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image 
```

```python
image1 = np.zeros((600,600,3))
cv2.circle(image1, (200,200), 200, color=(0,255,255), thickness=-1)
plt.imsave('../images/circle1.png', image1) 

image2 = np.zeros((600,600,3))
cv2.circle(image2, (300,300), 200, color=(0,255,255), thickness=-1)
plt.imsave('../images/circle2.png', image2) 
```

```python
i1 = cv2.imread('../images/circle1.png')
i2 = cv2.imread('../images/circle2.png')
plt.imshow(i1)
```

```python
plt.imshow(i2)
```

```python
circles_or = cv2.bitwise_or(i1, i2)
plt.imshow(circles_or)
```

```python
circles_and = cv2.bitwise_and(i1, i2)
plt.imshow(circles_and)
```

```python
circles_xor = cv2.bitwise_xor(i1, i2)
plt.imshow(circles_xor)
```

```python
circles_not = cv2.bitwise_not(i1)
plt.imshow(circles_not)
```
