```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
x1 = bd.drop(columns=['Price'])
y1 = bd['Price'] 
linmod3 = LinearRegression()
rfemod = RFE(linmod3,1)
rfemod.fit(x1,y1)
rfemod.ranking_
x1.columns

varlist = x1.columns
rating = rfemod.ranking_
pd.DataFrame([varlist,rating])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
linmod = LinearRegression()
x1 = bd.loc[:,['NOX','RM','CHAS','PTRATIO','DIS','LSTAT','RAD','CRIM','INDUS','ZN','TAX','B','AGE']]
y1 = bd['Price']
linmod.fit(x1,y1)
predPrice = linmod.predict(x1)
r2 = r2_score(y1,predPrice)

n = len(y1)    # Number of Samples
p = x1.shape[1] # Number of Input variables
adjr2 = 1 - (1-r2)*(n-1)/ (n-p-1)
print(r2,adjr2)


```
