```python
x = modfile.loc[:,['Sex_male','Pclass']]
y = modfile['Survived']
from sklearn.model_selection import train_test_split 
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=.25,random_state=1234,stratify=y)
```
