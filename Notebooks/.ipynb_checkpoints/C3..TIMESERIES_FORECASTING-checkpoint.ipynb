### Two major reasons for non-stationarity of a ts:


```python
'''
NOTES:

FORECASTING USING PYTHON STEPS:
1) Load Time Series data with date parser(.csv file)
3) Select time series(ts) from data frame #ts = data2['#Passengers']
3) Test Stationarity of a given time series
    a) Define test_stationarity(timeseries): 
        i)  Calculuate rolling mean & rolling std for 12 months periods:
                    --> timeseries.rolling(12).mean() & std()
        ii) Plot them against original time series
        ii) CalcPerform Dickey-Fuller Test 
    b) Test with log of our ts
        i)   Take the log of our ts: #ts_log = np.log(ts)
        ii)  Difference between original ts and ts_log
        iii) Test the difference # Result: 95% confidence it is stationary
    c) Test with exponentially weighted moving averge  #Result: 99% confidence that it is stationary
4) Apply other methods to remove trend and seasonality:
    1) Differencing (taking the difference with a particular time lag)
        i) ts_log_diff_x_minus_x1 = ts_log - ts_log.shift()
    2) Decomposition (modeling both trend and seasonality and removing them from model)
        i) statsmodels.tsa.seasonalseasonal_decompose(ts_log)
    
'''
```

### Step2: Test Stationarity of our time series

```python
from statsmodels.tsa.stattools import adfuller
```

#### DICKEY-FULLER TEST:    
    Stationary series: Test Statistic < Citical Values

#### def test_stationarity(timeseries)

```python
def test_stationarity(timeseries):
    ### Determine rolling statistics:################################
    rolling_mean = timeseries.rolling(12).mean() #(window=12)
    rolling_std = timeseries.rolling(12).std()
    
    print(type(rolling_mean)) #<class 'pandas.core.series.Series'>
    print(type(rolling_std))  # <class 'pandas.core.series.Series'>
    
    orginal = plt.plot(timeseries, color='blue', label='original')
    mean = plt.plot(rolling_mean, color='red', label='mean')
    std = plt.plot(rolling_std, color='black', label='std')
    plt.legend(loc='best') #(loc='upper left') 
    plt.title('Rolling MEAN & STD')
    plt.show(block=False)
    
    ###Perform Dickey-Fuller Test:################################
    df_test = adfuller(timeseries, autolag='AIC')    
    print('df_test:', '\n', df_test[:])
    print()  6y257354
    print('Results of Dickey-Fuller Test:')
    df_output = pd.Series(df_test[0:4], 
                          index=['Test Statistic', 'pValue', '#Lags used', 'Numer of observations used'])  
    for key, value in df_test[4].items():
        df_output['Critical Value (%s)'%key] = value
    print(df_output)
     
```

```python
test_stationarity(ts)
```

### Two major reasons for non-stationarity of a ts:
    1) Trend: varying mean over time (Ex: number of passenger growing over time)
    2) Seasonality: variations at specific time-frames (Ex: increase in car sales during holidays)

## Foreasting a Time Series

### ARIMA (stands for Auto-Regressive Integrated Moving Averages)

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
```

```python
lag_acf = acf(ts_log_diff_x_minus_x1, nlags=20)
lag_pacf = pacf(ts_log_diff_x_minus_x1, nlags=20, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff_x_minus_x1)),linestyle='--',color='blue')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff_x_minus_x1)),linestyle='--',color='pink')
plt.title('Autocorrelation Function')
```

```python
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff_x_minus_x1)),linestyle='--',color='blue')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff_x_minus_x1)),linestyle='--',color='pink')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

#### AR Model

```python
# set plot size
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
ar_model = ARIMA(ts_log, order=(2,1,0)) # p,d,q values as tuple 
```

```python
ar_results = ar_model.fit(disp=-1)
```

```python
plt.plot(ts_log_diff_x_minus_x1)
plt.plot(ar_results.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((ar_results.fittedvalues-ts_log_diff_x_minus_x1)**2))
```

#### MA Model

```python
ma_model = ARIMA(ts_log, order=(0,1,2)) # p,d,q values as tuple 
ma_results = ma_model.fit(disp=-1)
plt.plot(ts_log_diff_x_minus_x1)
plt.plot(ma_results.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((ma_results.fittedvalues-ts_log_diff_x_minus_x1)**2))
```

#### Combined Model

```python
combined_model = ARIMA(ts_log, order=(2,1,2)) # p,d,q values as tuple 
combined_results = combined_model.fit(disp=-1)
plt.plot(ts_log_diff_x_minus_x1)
plt.plot(combined_results.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((combined_results.fittedvalues-ts_log_diff_x_minus_x1)**2))
```

## Sales Forecasting using Spark (using Coudera's ARIMA model)
https://badrit.com/blog/2017/5/29/time-series-analysis-using-spark#.W3CfRpNKhTZ
 

```python
# This uses Cloudera version of ARIMA
Sample Code:
def predict_ARIMA_Spark(amounts, period):
    spark_context = pyspark.SparkContext.getOrCreate()
    model = spark_context._jvm.com.cloudera.sparkts.models.ARIMA.autoFit(_py2java(spark_context, Vectors.dense(amounts)), MAX_P, MAX_D, MAX_Q)
    p = _java2py(spark_context, model.p())
    d = _java2py(spark_context, model.d())
    q = _java2py(spark_context, model.q())
    jts = _py2java(spark_context, Vectors.dense(amounts))
    aic = model.approxAIC(jts)
    print "ARIMA(", p, d, q, ")", "AIC=", aic
    jfore = model.forecast(jts, period)
    return _java2py(spark_context, jfore)[len(amounts):]
```
